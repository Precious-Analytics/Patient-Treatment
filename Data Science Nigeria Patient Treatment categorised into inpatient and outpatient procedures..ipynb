## Need to scale the dataset

```python
rem_1 = test.drop(['patient_id_'], axis=1)
rem_1.head()

```

```python
rem = train[num_feat].drop(['SOURCE'], axis=1)
rem.head()
```

```python
rem.head()
```

```python


scaler = StandardScaler()
X = scaler.fit_transform(rem)
testing = scaler.transform(rem_1)

#X = pd.DataFrame(data=X, columns=cols)
X = pd.DataFrame(data=X, columns = rem.columns)
#testing = pd.DataFrame(data=testing, columns=cols)
testing = pd.DataFrame(data=testing, columns=rem_1.columns)

X.head()
```

## Building a base model

```python
X_train, X_test, y_train, y_test = train_test_split( X, train.SOURCE, test_size=0.30, random_state=42)
```

```python
classifier = SVC()

classifier.fit(X_train, y_train)
```

```python
pred = classifier.predict(X_test)

pred2 = classifier.predict(X_train)
```

```python
print('Training set')
print('classifier f1 score {}'.format(f1_score(y_train, pred2)))
print('Testing set')
print('classifier f1 score {}'.format(f1_score(y_test, pred)))
```
